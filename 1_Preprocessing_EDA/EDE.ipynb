This package lets us read from the .nii.gz files and access the array of the MRI

```python
import nibabel as nib
img = nib.load(test_path)
```

```python
#get_fdata accesses the array directly
image_data = img.get_fdata()

#size in pixels, (x,y,z)
image_data.shape
```

```python
image_data[:2]
```

```python
# 61 
len(image_data) # 61 
```

```python
# 71 
for i in image_data[:2]: 
    print(len(i))
```

```python
# 60
for i in image_data[:2]: 
    print(i.shape)
```

```python
# Can manually visualize one slice at a time
import matplotlib.pyplot as plt
```

```python
# Demo - slice shows a horizontal cross section 30 pixels up from the bottom of the brain (view from the top)
plt.imshow(image_data[:,:,30])
```

```python
# Plot each out of the 61 cross sections of the image
for i in range(image_data.shape[2]): 
    plt.imshow(image_data[:,:,i])
    plt.show()
```

```python
# Test - what are the other dimensions e.g. 0
for i in range(image_data.shape[0]): 
    plt.imshow(image_data[i,:,:])
    plt.show()
```

```python
# Test - what are the other dimensions e.g. 1
for i in range(image_data.shape[1]): 
    plt.imshow(image_data[:,i,:])
    plt.show()
```
